```python
import sqlite3
import numpy as np
import random
```

```python
conn = sqlite3.connect('testdb.sqlite')
c = conn.cursor()
c.execute('CREATE TABLE measurements(sensor_id integer, time integer, value real)')
conn.commit()
```

```python
c.execute('DELETE from measurements')
conn.commit()
for i in range(4):
    c.execute('INSERT into measurements(sensor_id,time, value) VALUES ('+str(i)+','+str(i*2)+','+str(i*3)+')')
```

```python
def get_meas_above(value):
    c.execute('SELECT sensor_id,time,value from measurements where value >='+str(value))
    return c.fetchall()

def get_meas_avg_value():
    c.execute('SELECT COUNT(sensor_id) FROM measurements UNION SELECT AVG(VALUE) from measurements')
    return c.fetchall()
print(get_meas_avg_value())

```
