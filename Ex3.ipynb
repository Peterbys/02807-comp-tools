```python

start = '2012-09-01'
end = '2012-10-01'
payload = {'start_at': start, 'end_at': end}
r = requests.get('https://api.exchangeratesapi.io/history',payload, headers = {'content-type': 'application/json'})
data = json.loads(r.text)
#print(json.dumps(data, indent=4))
json_normalize(data)
dates = [rate for rate in data['rates']]
dd = pd.DataFrame([])
for rate in data['rates']:
    dd = dd.append(pd.DataFrame([pd.Series(data['rates'][rate])]))
    #rate_vec = rate_vec.append(rate)
    #print(pd.DataFrame([pd.Series(data['rates'][rate])])
dd['dates'] = dates
#plt.plot(dd['dates'],dd['CAD'])
plt.plot(dd['dates'],dd['USD'])
```
